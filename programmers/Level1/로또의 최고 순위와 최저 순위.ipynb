ÎØºÏö∞Í∞Ä Íµ¨Îß§Ìïú Î°úÎòê Î≤àÌò∏Î•º Îã¥ÏùÄ Î∞∞Ïó¥ lottos, ÎãπÏ≤® Î≤àÌò∏Î•º Îã¥ÏùÄ Î∞∞Ïó¥ win_numsÍ∞Ä Îß§Í∞úÎ≥ÄÏàòÎ°ú Ï£ºÏñ¥ÏßëÎãàÎã§. Ïù¥Îïå, ÎãπÏ≤® Í∞ÄÎä•Ìïú ÏµúÍ≥† ÏàúÏúÑÏôÄ ÏµúÏ†Ä ÏàúÏúÑÎ•º Ï∞®Î°ÄÎåÄÎ°ú Î∞∞Ïó¥Ïóê Îã¥ÏïÑÏÑú return ÌïòÎèÑÎ°ù solution Ìï®ÏàòÎ•º ÏôÑÏÑ±Ìï¥Ï£ºÏÑ∏Ïöî.      

### ÏûÖÏ∂úÎ†• Ïòà
lottos	 |     win_nums	   | result              
[44, 1, 0, 0, 31, 25]	[31, 10, 45, 1, 6, 19]	[3, 5]       
[0, 0, 0, 0, 0, 0]	[38, 19, 20, 40, 15, 25]	[1, 6]       
[45, 4, 35, 20, 3, 9]	[20, 9, 3, 45, 4, 35]	[1, 1]        

```python
def solution(lottos, win_nums):
    answer = []
    lottos = sorted(lottos)
    win_nums = sorted(win_nums)
    zero = lottos.count(0)
    
    if zero == 6:
        return [1,6]
    
    win = 0
    for num in lottos[zero:]:
        if num in win_nums:
            win = win + 1
    
    if win == 0: # Îã§ ÌãÄÎ¶∞ Í≤ΩÏö∞ 
        return [6,6]
    else:
        max_ = win+zero
        return [6-max_+1, 6-win+1]
```

```python
lottos, win_nums = [44, 1, 0, 0, 31, 25], [31, 10, 45, 1, 6, 19]
print (solution(lottos, win_nums))

lottos, win_nums = [0, 0, 0, 0, 0, 0], [38, 19, 20, 40, 15, 25]
print (solution(lottos, win_nums))

lottos, win_nums = [45, 4, 35, 20, 3, 9], [20, 9, 3, 45, 4, 35]
print (solution(lottos, win_nums))
```

```python
# üëá ÌîÑÎ°úÍ∑∏ÎûòÎ®∏Ïä§ ÌíÄÏù¥ Ï§ë Í∞ÄÏû• ÍπîÎÅîÌïòÍ≥† Ï∂úÏ†ú ÏùòÎèÑÏóê ÎßûÍ≤å ÌíÄÏóàÎã§Í≥† ÏÉùÍ∞ÅÎêòÎäî ÏÜåÏä§ÏΩîÎìú
def solution(lottos, win_nums):

    rank=[6,6,5,4,3,2,1]

    cnt_0 = lottos.count(0)
    ans = 0
    for x in win_nums:
        if x in lottos:
            ans += 1
    return [rank[cnt_0 + ans],rank[ans]]
```

```python
lottos, win_nums = [44, 1, 0, 0, 31, 25], [31, 10, 45, 1, 6, 19]
print (solution(lottos, win_nums))

lottos, win_nums = [0, 0, 0, 0, 0, 0], [38, 19, 20, 40, 15, 25]
print (solution(lottos, win_nums))

lottos, win_nums = [45, 4, 35, 20, 3, 9], [20, 9, 3, 45, 4, 35]
print (solution(lottos, win_nums))
```
