[7, 0, 8, 2, 8, 3, 1, 5, 7, 6, 2]	"left"	"LRLLRRLLLRR"   
[1, 2, 3, 4, 5, 6, 7, 8, 9, 0]	"right"	"LLRLLRLLRL"   

```python
def solution(numbers, hand):
    answer = ''
    key = [[1,2,3],
           [4,5,6],
           [7,8,9],
            [-1,0,-2]]
    key_dic ={}
    
    for idx_r,row in enumerate (key):
        for idx,k in enumerate (row):
            key_dic[k] = [idx_r,idx]

    l_hand = key_dic[-1]
    r_hand = key_dic[-2]
    
    
    for num in numbers:
        if num == 1 or num == 4 or num == 7:
            answer += 'L'
            l_hand = key_dic[num]
            
        elif num == 3 or num == 6 or num == 9:
            answer += 'R'
            r_hand = key_dic[num]
            
        else:
            L_dis = sum([abs(L - now) for L, now in zip(l_hand, key_dic[num])])
            R_dis = sum([abs(R - now) for R, now in zip(r_hand, key_dic[num])])
            if L_dis < R_dis:
                answer += 'L'
                l_hand = key_dic[num]
            elif L_dis > R_dis: 
                answer += 'R'
                r_hand = key_dic[num]
            else: 
                if hand == 'left':
                    answer += 'L'
                    l_hand = key_dic[num]
                elif hand == 'right':
                    answer += 'R'
                    r_hand = key_dic[num]

    return answer
```
